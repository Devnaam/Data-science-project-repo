## Understand the data

### Subtask:
Get a general overview of the dataset, including the columns, data types, and a summary of the data.


**Reasoning**:
Load the dataset and get a general overview, including displaying the head, info, describe, checking for missing values and duplicates.



```python
import pandas as pd

df = pd.read_csv('/kaggle/input/netflix-shows/netflix_titles.csv')

print("First 5 rows of the dataset:")
display(df.head())

print("\nDataset Info:")
df.info()

print("\nStatistical Summary:")
display(df.describe())

print("\nMissing values per column:")
display(df.isnull().sum())

print("\nNumber of duplicate rows:")
print(df.duplicated().sum())
```

## Explore the data

### Subtask:
Explore the data by analyzing the distribution of categorical and numerical features, visualizing key aspects, and identifying trends and patterns.


**Reasoning**:
Analyze the distribution of categorical features, visualize key aspects, and identify trends and patterns as per the instructions. This involves analyzing the 'type', 'release_year', 'rating', 'country', and 'listed_in' columns.



```python
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter

# Step 1: Analyze the distribution of the 'type' column
plt.figure(figsize=(6, 4))
sns.countplot(x='type', data=df, palette='viridis')
plt.title('Distribution of Content Types (Movie vs. TV Show)')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

# Step 2: Examine the distribution of content over the years
plt.figure(figsize=(12, 6))
sns.histplot(data=df, x='release_year', bins=30, kde=True, color='skyblue')
plt.title('Distribution of Content Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()

# Step 3: Investigate the distribution of content ratings
plt.figure(figsize=(10, 6))
sns.countplot(x='rating', data=df, order=df['rating'].value_counts().index, palette='magma')
plt.title('Distribution of Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Step 4: Explore the top countries producing content
# Handle missing values in 'country' before splitting
df_country = df.dropna(subset=['country']).copy()
country_list = df_country['country'].str.split(', ').explode()
country_counts = Counter(country_list).most_common(10)
countries, counts = zip(*country_counts)

plt.figure(figsize=(12, 6))
sns.barplot(x=list(countries), y=list(counts), palette='plasma')
plt.title('Top 10 Countries Producing Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Step 5: Analyze the distribution of genres ('listed_in')
# Handle missing values in 'listed_in' before splitting
df_listed_in = df.dropna(subset=['listed_in']).copy()
genre_list = df_listed_in['listed_in'].str.replace(' ', '').str.split(',').explode()
genre_counts = Counter(genre_list).most_common(10)
genres, counts = zip(*genre_counts)

plt.figure(figsize=(12, 6))
sns.barplot(x=list(genres), y=list(counts), palette='viridis')
plt.title('Top 10 Content Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Clean the data

### Subtask:
Handle missing values, duplicates, and inconsistencies in the data.


**Reasoning**:
Identify columns with missing values and decide on an appropriate strategy for handling them based on the nature and number of missing values. Then implement the chosen strategy to handle missing values and verify that missing values have been handled. Finally, check for duplicate rows and remove them if they exist.



```python
# 1. Identify columns with missing values
print("Missing values before handling:")
display(df.isnull().sum())

# 2. & 3. Decide on and implement strategies for handling missing values
# 'director' and 'cast' have many missing values, replace with 'Unknown'
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')

# 'country', 'date_added', and 'rating' have fewer missing values, drop rows
df.dropna(subset=['country', 'date_added', 'rating'], inplace=True)

# 'duration' missing values are few, inspect and decide.
# Based on previous exploration, 'duration' is either minutes (Movies) or seasons (TV Shows).
# Let's inspect the remaining missing values in 'duration' after dropping rows with missing 'country', 'date_added', 'rating'.
print("\nMissing values in 'duration' after dropping rows:")
display(df['duration'].isnull().sum())

# If there are still missing values in 'duration', we can drop these rows as well.
df.dropna(subset=['duration'], inplace=True)


# 4. Verify that missing values have been handled
print("\nMissing values after handling:")
display(df.isnull().sum())

# 5. Confirm the absence of duplicate rows and remove if they exist
print("\nNumber of duplicate rows before dropping:")
print(df.duplicated().sum())

if df.duplicated().sum() > 0:
    df.drop_duplicates(inplace=True)
    print("\nNumber of duplicate rows after dropping:")
    print(df.duplicated().sum())
else:
    print("\nNo duplicate rows found.")

# 6. Address any inconsistencies (no explicit inconsistencies noted in previous steps, so this step is acknowledged but no code is needed based on previous findings)
```

## Feature engineering

### Subtask:
Create new features from existing ones that could be useful for analysis or modeling (e.g., extracting year or month from a date column).


**Reasoning**:
Create new features 'year_added', 'month_added', 'content_age', 'duration_value', and 'duration_unit' from the existing 'date_added', 'release_year', and 'duration' columns, handling potential errors and variations in format.



```python
import numpy as np

# 1. Convert 'date_added' to datetime objects. Handle potential errors by setting them to NaT
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

# 2. Extract the year the content was added
df['year_added'] = df['date_added'].dt.year

# 3. Extract the month the content was added
df['month_added'] = df['date_added'].dt.month

# 4. Calculate the age of the content when it was added. Handle potential errors or NaT values.
# Use the current year for calculation.
current_year = pd.Timestamp('now').year
df['content_age'] = current_year - df['release_year']

# For movies, duration is in minutes. For TV shows, it's seasons.
# 5. Separate duration value and unit
df['duration_value'] = df['duration'].str.extract('(\d+)').astype(float)
df['duration_unit'] = df['duration'].str.extract('([a-zA-Z]+)').astype(str)

# Verify the new columns
print("DataFrame with new features:")
display(df[['date_added', 'year_added', 'month_added', 'release_year', 'content_age', 'duration', 'duration_value', 'duration_unit']].head())

# Check for any NaNs created during feature engineering, especially for 'date_added' conversion
print("\nMissing values after feature engineering:")
display(df[['year_added', 'month_added', 'content_age', 'duration_value', 'duration_unit']].isnull().sum())
```

**Reasoning**:
Handle the remaining missing values in the newly created 'year_added' and 'month_added' columns which resulted from the 'date_added' column having missing values before conversion. Since these are derived features, and dropping rows with missing 'date_added' was already done, these remaining NaNs should correspond to those dropped rows, but the feature engineering step might have added them back if not careful. Re-checking and dropping rows with NaNs in 'year_added' and 'month_added' will ensure consistency.



```python
# Handle missing values in the newly created columns 'year_added' and 'month_added'.
# These NaNs likely correspond to the rows where 'date_added' was originally missing,
# which should have been dropped in the cleaning step.
# Let's re-drop rows where 'year_added' or 'month_added' are NaN to ensure data integrity.
df.dropna(subset=['year_added', 'month_added'], inplace=True)

# Verify that missing values have been handled for these columns
print("Missing values after handling NaNs in year_added and month_added:")
display(df[['year_added', 'month_added', 'content_age', 'duration_value', 'duration_unit']].isnull().sum())
```

## Ask specific questions and answer them

### Subtask:
Formulate specific questions about the data and use various analytical techniques to answer them.


**Reasoning**:
Identify the top 10 content genres by counting the occurrences of each genre in the 'listed_in' column and display the results.



```python
# 1. Identify the top 10 content genres
# Split the 'listed_in' column by comma and stack the values
genres_split = df['listed_in'].str.split(', ').explode()

# Count the occurrences of each genre
genre_counts = genres_split.value_counts()

# Get the top 10 genres
top_10_genres = genre_counts.head(10)

print("Top 10 Content Genres:")
display(top_10_genres)
```

**Reasoning**:
Determine how the amount of content added to Netflix has changed year over year by counting the number of titles added each year and display the results.



```python
# 2. Determine how the amount of content added to Netflix has changed year over year
# Count the number of titles added per year from the 'year_added' column
yearly_added_content = df['year_added'].value_counts().sort_index()

print("\nContent Added to Netflix Year Over Year:")
display(yearly_added_content)
```

**Reasoning**:
Find the distribution of content ratings by counting the occurrences of each rating in the 'rating' column, list the top 5 most common ratings, and display the results.



```python
# 3. Find the distribution of content ratings
rating_counts = df['rating'].value_counts()

# List the top 5 most common ratings
top_5_ratings = rating_counts.head(5)

print("\nDistribution of Content Ratings:")
display(rating_counts)

print("\nTop 5 Most Common Content Ratings:")
display(top_5_ratings)
```

**Reasoning**:
Identify the top 10 most frequent directors by counting the occurrences of each director in the 'director' column, excluding 'Unknown', and display the results.



```python
# 4. Identify the top 10 most frequent directors
# Split the 'director' column by comma and stack the values, excluding 'Unknown' directors
directors_split = df['director'].str.split(', ').explode()
known_directors = directors_split[directors_split != 'Unknown']

# Count the occurrences of each director
director_counts = known_directors.value_counts()

# Get the top 10 directors
top_10_directors = director_counts.head(10)

print("\nTop 10 Most Frequent Directors:")
display(top_10_directors)
```

**Reasoning**:
Identify the top 10 most frequent actors/actresses by counting the occurrences of each actor/actress in the 'cast' column, excluding 'Unknown', and display the results.



```python
# 5. Identify the top 10 most frequent actors/actresses
# Split the 'cast' column by comma and stack the values, excluding 'Unknown' cast
cast_split = df['cast'].str.split(', ').explode()
known_cast = cast_split[cast_split != 'Unknown']

# Count the occurrences of each actor/actress
cast_counts = known_cast.value_counts()

# Get the top 10 actors/actresses
top_10_cast = cast_counts.head(10)

print("\nTop 10 Most Frequent Actors/Actresses:")
display(top_10_cast)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Distribution of Content Types
plt.figure(figsize=(6, 4))
sns.countplot(x='type', data=df, palette='viridis')
plt.title('Distribution of Content Types')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()

# 2. Content Added Over the Years
plt.figure(figsize=(12, 6))
yearly_added_content.plot(kind='line', marker='o', color='skyblue')
plt.title('Content Added to Netflix Year Over Year')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles Added')
plt.grid(True)
plt.show()

# 3. Distribution of Content Ratings
plt.figure(figsize=(10, 6))
sns.barplot(x=rating_counts.index, y=rating_counts.values, palette='magma')
plt.title('Distribution of Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 4. Top 10 Countries Producing Content
plt.figure(figsize=(12, 6))
sns.barplot(x=list(countries), y=list(counts), palette='plasma')
plt.title('Top 10 Countries Producing Content')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 5. Top 10 Content Genres
plt.figure(figsize=(12, 6))
sns.barplot(x=list(genres), y=list(counts), palette='viridis')
plt.title('Top 10 Content Genres')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 6. Top 10 Most Frequent Directors
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_directors.index, y=top_10_directors.values, palette='rocket')
plt.title('Top 10 Most Frequent Directors')
plt.xlabel('Director')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# 7. Top 10 Most Frequent Actors/Actresses
plt.figure(figsize=(12, 6))
sns.barplot(x=top_10_cast.index, y=top_10_cast.values, palette='mako')
plt.title('Top 10 Most Frequent Actors/Actresses')
plt.xlabel('Actor/Actress')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
